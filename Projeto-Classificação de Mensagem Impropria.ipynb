```python
def registrar_elogio(mensagem, elogios):

    with open("elogios.txt", "a") as arquivo_elogio:
      for palavra in mensagem.split():
        if palavra.lower() in elogios:
            arquivo_elogio.write(palavra.lower() + "\n")

def registrar_reclamacao(mensagem, reclamacoes):
    with open("reclamacoes.txt", "a") as arquivo_reclamacao:
      for palavra in mensagem.split():
        if palavra.lower() in reclamacoes:
            arquivo_reclamacao.write(palavra.lower() + "\n")

def registrar_palavra_impropria(mensagem, palavras_improprias):
    with open("palavras_improprias.txt", "a") as arquivo_palavra_impropria:
      for palavra in mensagem.split():
        if palavra.lower() in palavras_improprias:
          arquivo_palavra_impropria.write(palavra.lower() + "\n")

def verificar_mensagem(mensagem):
    palavras_improprias = ["merda", "porra", "caralho",
    "merda",
    "puta",
    "foda",
    "porra",
    "bosta",
    "cuzão",
    "viado",
    "vadia",
    "fudeu",
    "fudido",
    "cacete",
    "buceta",
    "cu",
    "vai se foder",
    "foda-se",
    "pentelho",
    "pau no cu",
    "filho da puta",
    "arrombado"
]
    elogios = ["perfeito", "bom", "gostei", "atrativo", "arrojado", "assertivo", "atencioso","Parabéns",
    "Excelente",
    "Brilhante",
    "Fantástico",
    "Incrível",
    "Magnífico",
    "Maravilhoso",
    "Espetacular",
    "Excepcional",
    "Extraordinário",
    "Notável",
    "Exemplar",
    "Genial",
    "Estupendo",
    "Fenomenal",
    "Impressionante",
    "Admirável",
    "Sensacional",
    "Surpreendente",
    "Esplêndido"]
    reclamacoes = ["ruim", "fosse", "melhore", "deveria", "pessimo","insatisfeito",
    "problema",
    "atendimento ruim",
    "atraso",
    "falha",
    "incompetência",
    "desapontamento",
    "insatisfação",
    "insuficiente",
    "erro",
    "inconveniente",
    "inadequado",
    "descontente",
    "ineficiente",
    "descaso",
    "imprecisão",
    "insuficiência",
    "inconformidade",
    "incompleto",
    "negligência"]

    if any(palavra in mensagem for palavra in palavras_improprias):
        registrar_palavra_impropria(mensagem, palavras_improprias)
    elif any(palavra in mensagem for palavra in elogios):
        registrar_elogio(mensagem, elogios)
    elif any(palavra in mensagem for palavra in reclamacoes):
        registrar_reclamacao(mensagem, reclamacoes)
    else:
        print("Mensagem não classificada.")

def main():
    while True:
        print("Digite sua mensagem ou 'sair' para encerrar:")
        mensagem = input("> ")

        if mensagem.lower() == "sair":
            print("Encerrando o programa.")
            break

        verificar_mensagem(mensagem)

if __name__ == "__main__":
    main()
```
